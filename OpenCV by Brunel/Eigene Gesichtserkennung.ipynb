```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#Ohne matplotlib
img = cv2.imread("4112d43fef464f46a1d68b0de05d8369.jpg")
img.shape
cv2.imshow('image 1', img)

cv2.waitKey(0)
```

```python
#mit matplotlib
img = cv2.imread("4112d43fef464f46a1d68b0de05d8369.jpg")
bru = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
bru.shape
plt.imshow(bru)
plt.show()
```

```python
#Ohne matplotlib
classifier = cv2.CascadeClassifier(r"C:\Users\amour\Desktop\Python\Kursmaterialen\Kursmaterialien\data\haarcascades\haarcascade_frontalface_alt2.xml")
faces = classifier.detectMultiScale(img, minNeighbors = 5)

img_1 = img.copy()
for face in faces:
    x,y,w,h = face
    cv2.rectangle(img_1, (x, y), (x + w, y + h), (0, 255, 0), 3)
    print(face)
cv2.imshow('image 2', img_1)

cv2.waitKey(0)
```
