```python
# Bibliothek-system
class Book():
    def __init__(self,title = "unknown",author = "unknown",editor = "unknown",year = "unknown"):
        self.title = title
        self.author = author
        self.editor = editor
        self.year = year

    def __str__(self):
        return f"Title: {self.title} \nAuthor: {self.author} \nEditor: {self.editor} \nYear of publication: {self.year} \n"

class Library():
    def __init__(self):
        self.list_book = []

    def add_book(self,*books):
        for book in books:
            self.list_book.append(book)
            print("Book sucessfully added in the Library. Here are the Informations: ")
            print(f"Title: {book.title} \nAuthor: {book.author} \nEditor: {book.editor} \nYear of publication: {book.year} \n")
        
    def remove_book(self,*books):
        for book in books:
            if book in self.list_book:
                self.list_book.remove(book)
                print(f"The book \"{book.title}\" has been removed \n")
            else:
                print("This Book is not here.\n")

    def search_book(self,*books):
        for book in books:
            if book in self.list_book:
                print(f"The book \"{book.title}\" ist here \n")
            else:
                print(f"We don't have that book \n")

class User():
    def __init__(self,name):
        self.name = name
        self.loan_list = []

    def loan_book(self,book,library):
        if book not in self.loan_list:
            library.list_book.remove(book)
            self.loan_list.append(book)
            print(f"You borrowed the book \"{book.title}\". Please return the Book when you finish with it.\n")
        else:
            print("You already borrowed this book.\n")
    def give_back_book(self,book,library):
        if book in self.loan_list:
            self.loan_list.remove(book)
            library.list_book.append(book)
            print(f"You gave back the book \"{book.title}\".\n")
        else:
            print("You didn't borrowed this book.\n")
            
book_1 = Book("Les Tribus de Capitoline","PC Ombete Mbella","BCS",2009)
book_2 = Book("Munyal","Djaïli Amadou Amal","Afric-editon",2015)
book_3 = Book()

b = Library()
b.add_book(book_1,book_2,book_3)
b.search_book(book_1,book_2)
print(book_1)
print(book_2)

user_1 = User("Gabriel")
user_1.loan_book(book_1,b)

b.search_book(book_1)
```
